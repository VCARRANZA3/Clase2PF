```python
#Librerias utilizadas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Preparativos para la base
data_students = pd.read_csv("clean_students_complete.csv")
data_students.set_index("Student ID", inplace= True)
data_students = data_students.drop(["Unnamed: 0"], axis=1)
data_students
```

```python
#Convertir a variable categoricas
data_students["cat_read"]=data_students["reading_score"]
data_students["cat_math"]=data_students["math_score"]
data_students['cat_read']=pd.Categorical(data_students.cat_read)
data_students['cat_math']=pd.Categorical(data_students.cat_math)
data_students["cat_read"].replace([range(70,100)], "Aprobado", inplace=True)
data_students["cat_read"].replace([range(0,70)], "Reprobado", inplace=True)
data_students["cat_math"].replace([range(70,100)], "Aprobado", inplace=True)
data_students["cat_math"].replace([range(0,70)], "Reprobado", inplace=True)
data_students
```

```python
#Grafica de aprobados-reprobados de cat_read
resumen1=data_students[["school_name","cat_read"]]
dummies1=pd.get_dummies(resumen1[["cat_read"]])
base1=pd.concat([resumen1,dummies1],axis=1)
base1=base1.groupby(["school_name"]).sum()
base1.plot(kind="bar",color=["red","green"],figsize=(13,9))
```

```python
#Grafica de aprobados-reprobados de cat_math
resumen2=data_students[["school_name","cat_math"]]
dummies2=pd.get_dummies(resumen2[["cat_math"]])
base2=pd.concat([resumen2,dummies2],axis=1)
base2=base2.groupby(["school_name"]).sum()
base2.plot(kind="bar",color=["red","green"],figsize=(13,9))
```
