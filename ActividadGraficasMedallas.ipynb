```python
#Crea una gráfica de barras que represente a los 10 países con más medallas ganadas
import pandas as pd
import numpy as np
Atletas = pd.read_csv("athlete_events.csv", index_col=0) #Subo el archivo
Campeones = Atletas[Atletas["Medal"].notna()] #Queda solo medallistas
Paises = pd.DataFrame(Campeones["Team"].value_counts()) #Conteo de medallas por país
Top10 = Paises.head(10) #Los mejores 10
Top10["Team"].plot(kind="bar") #Grafica
```
