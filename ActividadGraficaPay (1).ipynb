Actividad Grafica Pay de Victor Carranza

```python
import pandas as pd
import numpy as np
Atletas = pd.read_csv("athlete_events.csv", index_col=0) #Subo el archivo
Campeones = Atletas[Atletas["Medal"].notna()] #Queda solo medallistas
Tiempo = Campeones[["Year","Season","Medal","NOC"]] #Base mas pequeña
Desglose = pd.get_dummies(Tiempo[["Medal"]]) #Dummies para medallas
Unido = pd.concat([Tiempo,Desglose], axis=1) #Unir las bases
Tabla = Unido.groupby(["Year","Medal","Season"]).sum() #Juntar por agrupacion
Tabla["Total_Medals"] = Tabla["Medal_Bronze"] + Tabla["Medal_Gold"] + Tabla["Medal_Silver"] #Crear columna total
Reciente = Tabla[150:153] #Tener ultimo año
Nombres = ["Bronze","Gold","Silver"] #Agregar las etiquetas
Grafica = Reciente["Total_Medals"].plot.pie(figsize=(11,6),labels=Nombres,autopct="%0.1f %%") #Generar la grafica
Imagen = Grafica.get_figure()
Imagen.savefig("Grafica Pay.png")
```
