```python
#Librerias utilizadas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Preparativos para la base
data_students = pd.read_csv("clean_students_complete.csv")
data_students.set_index("Student ID", inplace= True)
data_students = data_students.drop(["Unnamed: 0"], axis=1)
data_students
```

```python
#Proceso para el mejor puntaje por escuela y grado
max_reading=data_students[data_students["reading_score"]==data_students["reading_score"].max()]
max_reading_corto=max_reading[["gender","school_name","grade"]]
dummies1=pd.get_dummies(max_reading_corto[["gender"]])
max_reading_base=pd.concat([max_reading_corto,dummies1],axis=1)
max_reading_base=max_reading_base.groupby(["grade","school_name"]).sum()
```

```python
#Grafica de noveno grado
max_reading_9th=max_reading_base[45:60]
max_reading_9th.plot(kind="bar",color=["red","blue"],figsize=(13,9))
```

```python
#Grafica de decimo grado
max_reading_10th=max_reading_base[0:15]
max_reading_10th.plot(kind="bar",color=["red","blue"],figsize=(13,9))
```

```python
#Grafica de onceavo grado
max_reading_11th=max_reading_base[15:30]
max_reading_11th.plot(kind="bar",color=["red","blue"],figsize=(13,9))
```

```python
#Grafica de doceavo grado
max_reading_12th=max_reading_base[30:45]
max_reading_12th.plot(kind="bar",color=["red","blue"],figsize=(13,9))
```
