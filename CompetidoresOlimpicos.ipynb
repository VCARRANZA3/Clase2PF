```python
#Crear un programa que obtenga el competidor m치s veterano que ha recibido medalla
Atleta = pd.read_csv("athlete_events.csv", index_col=0)
Campeones = Atleta[Atleta["Medal"].notna()]
Campeones["Age"].max() #Aqui dio 73
CampeonVeterano = Campeones[Campeones["Age"]==73]
CampeonVeterano
```

```python
#Crear un programa que obtenga el competidor m치s joven que ha recibido medalla de oro
CampeonesOro = Atleta[Atleta["Medal"]=="Gold"]
CampeonesOro["Age"].min() #Aqui dio 13
CampeonOroJoven = CampeonesOro[CampeonesOro["Age"]==13]
CampeonOroJoven
```

```python
#Encuentra al competidor m치s ganador de la historia y crea un csv con toda su informaci칩n.
Campeones["Name"].value_counts() #Aqui dio Michael Phelps
MaximoCampeon = Campeones[Campeones["Name"]=="Michael Fred Phelps, II"]
MaximoCampeon.to_csv("MaximoCampeon.csv", header=True, index=True)
MaximoCampeon
```
